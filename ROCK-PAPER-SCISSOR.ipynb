```python
import random

# Print the rules of the game
print("Welcome to the ROCK PAPER SCISSORS Showdown!\n"
      + "Here are the rules:\n"
      + "🪨 Rock vs 📜 Paper -> 📜 Paper wraps 🪨 Rock (Paper wins)\n"
      + "🪨 Rock vs ✂️ Scissors -> 🪨 Rock crushes ✂️ Scissors (Rock wins)\n"
      + "📜 Paper vs ✂️ Scissors -> ✂️ Scissors cuts 📜 Paper (Scissors wins)\n")

while True:
    print("\nMake your move!\n 1 - 🪨 Rock \n 2 - 📜 Paper \n 3 - ✂️ Scissors \n")

    # Take the input from user
    choice = int(input("Enter your choice (1/2/3): "))

    # Loop until user enters a valid choice
    while choice > 3 or choice < 1:
        choice = int(input("That's not a valid move! Try again (1/2/3): "))

    # Map the choice to the corresponding name
    if choice == 1:
        choice_name = '🪨 Rock'
    elif choice == 2:
        choice_name = '📜 Paper'
    else:
        choice_name = '✂️ Scissors'

    # Print user's choice
    print(f"\nYou chose: {choice_name} 🆚 Now, it's the computer's turn...")

    # Computer randomly chooses its move
    comp_choice = random.randint(1, 3)

    # Ensure the computer choice is different from the user's choice
    while comp_choice == choice:
        comp_choice = random.randint(1, 3)

    # Map the computer's choice to the corresponding name
    if comp_choice == 1:
        comp_choice_name = '🪨 Rock'
    elif comp_choice == 2:
        comp_choice_name = '📜 Paper'
    else:
        comp_choice_name = '✂️ Scissors'

    # Print computer's choice
    print(f"The computer chose: {comp_choice_name}\n")
    print(f"🏁 Showdown: {choice_name} 🆚 {comp_choice_name}")

    # Determine the result
    if choice == comp_choice:
        result = "DRAW"
        print("It's a stalemate! Both you and the computer chose the same move. 🤝")
    elif (choice == 1 and comp_choice == 2) or (choice == 2 and comp_choice == 1):
        result = '📜 Paper'
        print(f"📜 Paper covers 🪨 Rock! {result} wins!")
    elif (choice == 1 and comp_choice == 3) or (choice == 3 and comp_choice == 1):
        result = '🪨 Rock'
        print(f"🪨 Rock smashes ✂️ Scissors! {result} wins!")
    elif (choice == 2 and comp_choice == 3) or (choice == 3 and comp_choice == 2):
        result = '✂️ Scissors'
        print(f"✂️ Scissors cut 📜 Paper! {result} wins!")

    # Announce the final outcome
    if result == choice_name:
        print("🎉 Congratulations, you won! 🏆")
    elif result == "DRAW":
        print("It's a tie! Well played. 🥈")
    else:
        print("Oh no, the computer won this round! Better luck next time. 💻")

    # Ask the user if they want to play again
    play_again = input("\nWould you like to play another round? (Y/N): ").lower()
    if play_again == 'n':
        break

# After exiting the loop
print("\nThanks for playing ROCK PAPER SCISSORS! Come back soon!")
```
